## 1.1 O que √© An√°lise Explorat√≥ria de Dados (EDA)?

De forma simples, **An√°lise Explorat√≥ria de Dados (EDA)** √© o conjunto de t√©cnicas e ferramentas usadas para:

- **Explorar** um conjunto de dados;
- **Descrever** o que est√° acontecendo;
- **Entender** padr√µes, erros, problemas e rela√ß√µes entre vari√°veis.

√â como se voc√™ estivesse conhecendo uma pessoa nova: antes de fazer qualquer julgamento ou tomar decis√µes importantes, voc√™ faz perguntas, observa comportamentos, presta aten√ß√£o em detalhes.  
Com dados √© a mesma coisa: antes de construir modelos, previs√µes ou tomar decis√µes com base neles, precisamos **entender que dados s√£o esses**.

Alguns objetivos t√≠picos da EDA:

- Ver como os valores est√£o distribu√≠dos (tem muita gente com nota alta? baixa? tudo misturado?);
- Identificar valores faltantes (campos vazios, ‚ÄúNaN‚Äù, ‚ÄúNone‚Äù);
- Detectar valores estranhos ou extremos (**outliers**), que fogem muito do padr√£o;
- Resumir os dados com estat√≠sticas como m√©dia, mediana, desvio padr√£o;
- Visualizar rela√ß√µes entre vari√°veis (por exemplo: ‚Äúquem estuda mais tira nota maior?‚Äù).

> Um ponto importante: a EDA **vem antes** de criar modelos preditivos (como regress√µes, redes neurais, etc.).  
> Se voc√™ tentar modelar sem explorar os dados, √© como montar um quebra-cabe√ßa sem olhar a figura da caixa.

Neste mini-curso, o nosso foco √© exatamente esse ‚Äúprimeiro contato inteligente‚Äù com os dados: aprender a fazer perguntas para o dataset, olhar para as respostas com senso cr√≠tico e evitar armadilhas comuns.

## 3.2. Por que n√£o fazer tudo "na m√£o"?

Em teoria, poder√≠amos fazer muitas coisas apenas com o Python puro, por exemplo:

- Calcular m√©dias usando listas e la√ßos `for`;
- Guardar dados em listas de listas (como uma "tabela" manual).

Mas isso seria:

- Mais **trabalho**;
- Mais f√°cil de **errar**;
- Mais dif√≠cil de **reutilizar** e **organizar**.

As bibliotecas como `pandas` e `numpy` j√° trazem:

- Fun√ß√µes prontas para ler arquivos `.csv`;
- Estruturas otimizadas para trabalhar com tabelas e n√∫meros;
- Opera√ß√µes r√°pidas em colunas inteiras (sem precisar de `for`).

Vamos ver um pequeno exemplo comparando uma m√©dia ‚Äúna m√£o‚Äù com a m√©dia usando `numpy`.

### C√≥digo de Exemplo 2 - M√©dia na m√£o vs. m√©dia com numpy

````python
# Exemplo: calculando a m√©dia de uma lista de notas
grades = [5.0, 7.5, 8.0, 9.0, 6.5]
````

````python
# 1) M√©dia "na m√£o" (Python puro)
sum_grades = 0
for grade in grades:
    sum_grades += grade

manual_mean = sum_grades / len(grades)
````

````python
# 2) M√©dia usando numpy
np_mean = np.mean(grades)
````

````python
print("Notas:", grades)
print("M√©dia calculada na m√£o:", manual_mean)
print("M√©dia calculada com numpy:", np_mean)
````

### C√≥digo de Pr√°tica 2 - Criando um DataFrame

Vamos criar um DataFrame "na m√£o", para testar se o `pandas` est√° funcionando de forma correta.

Al√©m disso, para os que n√£o tem familiaridade, ir√£o entender um pouco mais como funciona essa estrutura de dados.

Exemplo Did√°tico: O Dicion√°rio 'Aluno'

````python
# 1. O que √© um Dicion√°rio?
# Um dicion√°rio √© uma cole√ß√£o de itens que armazena dados em pares de 'chave: valor'.
# Pense nele como um "telefone" onde cada nome (chave) tem um n√∫mero (valor) associado.
# As chaves s√£o √∫nicas e podem ser strings ou n√∫meros, enquanto os valores podem ser de qualquer tipo.

# Vamos criar um dicion√°rio para representar um 'aluno':
aluno = {
    "nome": "Maria Silva",
    "idade": 16,
    "serie": "2¬∫ ano",
    "notas": [7.5, 8.0, 9.2],
    "aprovado": True
}

print("\n1. Nosso dicion√°rio 'aluno' inicial:")
print(aluno)
print(f"\nTipo do 'aluno': {type(aluno)}")
````

````python
# 2. Acessando valores no dicion√°rio
# Para pegar um valor, usamos a chave dentro de colchetes, como em aluno["chave"].

print("\n2. Acessando informa√ß√µes espec√≠ficas:")
print(f"Nome do aluno: {aluno['nome']}")
print(f"Idade do aluno: {aluno['idade']} anos")
print(f"Notas do aluno: {aluno['notas']}")
````

````python
# 3. Adicionando uma nova informa√ß√£o
# Para adicionar um novo par chave-valor, basta atribuir um valor a uma nova chave.

aluno["cidade"] = "S√£o Paulo"
print("\n3. Dicion√°rio ap√≥s adicionar a cidade:")
print(aluno)
````

````python
# 4. Modificando uma informa√ß√£o existente
# Para mudar um valor, basta atribuir um novo valor a uma chave existente.

aluno["idade"] = 17
print("\n4. Dicion√°rio ap√≥s modificar a idade:")
print(aluno)
````

````python
# 5. Verificando as chaves e valores
print("\n5. Todas as chaves do dicion√°rio:")
print(aluno.keys())

print("\n6. Todos os valores do dicion√°rio:")
print(aluno.values())

print("\n7. Todos os itens (chave:valor) do dicion√°rio:")
print(aluno.items())
````

Criando um DataFrame a partir de um Dicion√°rio.

## 5.8. Agrupamento Simples

O m√©todo `groupby` permite agrupar os dados por uma coluna categ√≥rica e calcular m√©tricas em uma coluna num√©rica.

Exemplos:

- M√©dia de nota por curso;
- M√©dia de horas de estudo por hor√°rio preferido;
- Quantidade de alunos por combina√ß√£o de categorias.

Sintaxe b√°sica:


- `df.groupby("coluna_agrupada")["coluna_avaliada"].funcao()`


Fun√ß√µes que geralmente s√£o utilizadas:

| Fun√ß√£o | Descri√ß√£o |
| :--- | :--- |
| **`sum()`** | Calcula a **soma** dos valores em cada grupo. |
| **`mean()`** | Calcula a **m√©dia aritm√©tica** dos valores em cada grupo. |
| **`median()`** | Calcula a **mediana** (valor central) dos valores em cada grupo. |
| **`min()`** | Encontra o **valor m√≠nimo** em cada grupo. |
| **`max()`** | Encontra o **valor m√°ximo** em cada grupo. |
| **`count()`** | Conta o **n√∫mero de valores n√£o-nulos** em cada grupo. |
| **`size()`** | Conta o **n√∫mero total de linhas** (tamanho) em cada grupo (incluindo nulos). |
| **`std()`** | Calcula o **desvio-padr√£o** dos valores em cada grupo. |
| **`var()`** | Calcula a **vari√¢ncia** dos valores em cada grupo. |
| **`first()`** | Retorna o **primeiro valor** de cada grupo. |
| **`last()`** | Retorna o **√∫ltimo valor** de cada grupo. |

### C√≥digo de Exemplo 8 -

````python
# M√©dia das Notas por Curso
mean_score_by_course = df_demo.groupby("course")["exam_score"].mean()
mean_score_by_course
````

## 6.4. O Quarteto de Ascombe

````python
# Op√ß√£o A: Criando manualmente (recomendado para transpar√™ncia)
data_anscombe = {
    'x1': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y1': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68],
    'x2': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y2': [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74],
    'x3': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y3': [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73],
    'x4': [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8],
    'y4': [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
}

df_anscombe = pd.DataFrame(data_anscombe)
````

````python
# Pe√ßa para calcular a m√©dia e desvio padr√£o de x1 e x2, y1 e y2. Ser√£o quase id√™nticos.
print("Dataset Anscombe Carregado!")
display(df_anscombe.head())
````

````python
# Lista para armazenar as m√©tricas de cada grupo
resultados = []

# Iterar sobre os 4 grupos (1 a 4)
for i in range(1, 5):
    # Selecionar as colunas corretas (ex: x1 e y1, x2 e y2...)
    col_x = f'x{i}'
    col_y = f'y{i}'

    x = df_anscombe[col_x]
    y = df_anscombe[col_y]

    # Calcular Regress√£o Linear (y = ax + b) para pegar declividade e intercepto
    coeficiente_angular, intercepto = np.polyfit(x, y, 1)

    # Guardar as m√©tricas num dicion√°rio
    resultados.append({
        'Dataset': f'Grupo {i}',
        'M√©dia X': x.mean(),
        'M√©dia Y': y.mean(),
        'Desvio Padr√£o X': x.std(),
        'Desvio Padr√£o Y': y.std(),
        'Correla√ß√£o': x.corr(y),
        'Linha de Regress√£o': f'y = {coeficiente_angular:.2f}x + {intercepto:.2f}'
    })

# Criar um DataFrame para visualiza√ß√£o limpa
df_metricas = pd.DataFrame(resultados)

# Exibir a tabela
print("--- Compara√ß√£o das Estat√≠sticas Descritivas ---")
display(df_metricas.round(2))
# Dica de aula: Destaque que at√© a equa√ß√£o da reta √© quase id√™ntica em todos!
````

### Grupo 1: O "Mundo Ideal" (Rela√ß√£o Linear Simples)

Este √© o gr√°fico onde a estat√≠stica cl√°ssica funciona bem.
- **O Gr√°fico:** Uma nuvem de pontos que segue uma linha ascendente com um pouco de dispers√£o natural.
- **O Caso Real:** Pre√ßo de Im√≥veis vs. Metragem Quadrada.
- **A Narrativa:** Em geral, quanto maior o apartamento, mais caro ele √©. Claro, existem varia√ß√µes (um apartamento velho pode ser mais barato, um reformado mais caro), mas a regra geral se aplica: aumentou $x$, aumentou $y$.
- **Conclus√£o:** Aqui, a m√©dia e a correla√ß√£o descrevem bem a realidade. Pode confiar no n√∫mero.

### Grupo 2: A "Lei dos Rendimentos Decrescentes" (Rela√ß√£o N√£o-Linear)

Este √© o gr√°fico da curva (par√°bola). A linha de regress√£o diz que √© uma reta, mas visivelmente n√£o √©.

- **O Gr√°fico:** Uma curva que sobe e depois desce (ou desacelera).
- **O Caso Real:** Produtividade vs. Horas Trabalhadas.
- **A Narrativa:** Se voc√™ trabalha 2 horas, produz X. Se trabalha 8 horas, produz 4X. A estat√≠stica linear diz: "Ent√£o se voc√™ trabalhar 24 horas seguidas, produzir√° 12X!".
  - **A Realidade:** N√£o. Depois de 10 ou 12 horas, voc√™ fica exausto, comete erros e a produtividade cai drasticamente. A rela√ß√£o n√£o √© uma reta infinita, ela tem um teto f√≠sico.
- **O Perigo:** Se voc√™ usar apenas a m√©dia/regress√£o linear aqui, vai tomar decis√µes desastrosas (como obrigar a equipe a fazer hora extra achando que a produ√ß√£o vai subir linearmente).

### Grupo 3: A "Exce√ß√£o que Estraga a Regra" (Outlier de Regress√£o)

Este √© o gr√°fico da linha perfeita com um √∫nico ponto fora.

- **O Gr√°fico:** Todos os pontos perfeitamente alinhados, exceto um que est√° longe da linha.
- **O Caso Real:** Calibra√ß√£o de Sensores ou Erro de Digita√ß√£o.
- **A Narrativa:** Imagine um term√¥metro industrial medindo a temperatura de um forno. O forno sobe 1 grau por minuto perfeitamente.
  - Minuto 1: 100¬∞C
  - Minuto 2: 101¬∞C
  - Minuto 3: 500¬∞C (Erro! O sensor falhou ou algu√©m digitou um zero a mais).
  - Minuto 4: 103¬∞C.
- **O Perigo:** A correla√ß√£o vai cair e a m√©dia vai subir absurdamente por causa de um erro. O processo √© perfeito, a coleta de dados √© que falhou. A estat√≠stica descritiva diz que o forno √© "inst√°vel", mas o gr√°fico mostra que foi apenas um glitch.

### Grupo 4: A "Fal√°cia da Correla√ß√£o" (Ponto de Alavancagem)

Este √© o mais bizarro: todos os $x$ s√£o iguais (uma linha vertical), exceto um ponto l√° longe que faz parecer que existe uma linha inclinada.
- **O Gr√°fico:** Uma parede vertical de pontos e um ponto solit√°rio longe dela.
- **O Caso Real:** Instala√ß√£o de Vers√£o de Software (Bin√°rio) vs. Erros.
- **A Narrativa:** Imagine que voc√™ est√° analisando bugs em vers√µes de um sistema.
  - 10 computadores t√™m a Vers√£o 1.0 instalada (X=1). Todos tiveram 0 erros.
  - 1 computador (do desenvolvedor) tem a Vers√£o 2.0 (X=2) e teve 10 erros.
  - Se voc√™ rodar uma regress√£o linear, o computador vai tra√ßar uma linha ligando a Vers√£o 1 √† Vers√£o 2 e dizer: "Existe uma tend√™ncia de crescimento de erros".
- **O Perigo:** Voc√™ n√£o tem uma tend√™ncia real, voc√™ tem dois estados diferentes (Vers√£o 1 e Vers√£o 2). N√£o existe "Vers√£o 1.5". Esse ponto solit√°rio (leverage point) cria a ilus√£o de uma correla√ß√£o cont√≠nua onde, na verdade, voc√™ deveria estar comparando categorias (Grupo A vs Grupo B), n√£o fazendo uma regress√£o linear.

## 8.3. Regra Simples Para Detec√ß√£o De Outliers (IQR)  

Uma regra muito usada para detectar outliers √© baseada no **IQR (Interquartile Range)**:

- Q1 = primeiro quartil (25% dos dados abaixo desse valor);
- Q3 = terceiro quartil (75% dos dados abaixo desse valor);
- IQR = Q3 - Q1.

Definimos limites "esperados":

- **Limite inferior** = Q1 - 1.5 √ó IQR  
- **Limite superior** = Q3 + 1.5 √ó IQR  

Valores abaixo do limite inferior ou acima do limite superior s√£o considerados **poss√≠veis outliers**.

Essa regra n√£o √© perfeita, mas funciona bem como aproxima√ß√£o em muitos casos.

### C√≥digo de Exemplo 3

````python
# Calculando Q1, Q3, IQR e limites para study_hours_per_day
Q1 = df_outliers_demo["study_hours_per_day"].quantile(0.25)
Q3 = df_outliers_demo["study_hours_per_day"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
````

````python
# Exibindo os valores calculados
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Limite inferior:", lower_bound)
print("Limite superior:", upper_bound)
````

````python
# Identificando poss√≠veis outliers
outlier_mask = (df_outliers_demo["study_hours_per_day"] < lower_bound) | (df_outliers_demo["study_hours_per_day"] > upper_bound)
outliers = df_outliers_demo[outlier_mask]
outliers
````

## 8.4. O que fazer com Outliers?

Depois de detectar outliers, vem a parte mais delicada: **decidir o que fazer com eles**.

Alguns cen√°rios comuns:

1. **Erro de digita√ß√£o/registro**
   - Ex.: nota = 1000 em uma prova que vai at√© 100.
   - A√ß√£o t√≠pica: corrigir o valor (se souber o correto) ou remover o registro.

2. **Caso raro, mas real**
   - Ex.: um aluno que realmente estudou muito mais do que o resto.
   - A√ß√£o: muitas vezes vale a pena **manter**, mas destacar que ele √© um caso especial.

3. **Valor plaus√≠vel, mas que distorce muito as estat√≠sticas**
   - Ex.: uma lat√™ncia de servidor muito alta em um dia de falha.
   - A√ß√µes poss√≠veis:
     - Remover para n√£o "puxar" a m√©dia de todo o per√≠odo;
     - Truncar (limitar) o valor a um m√°ximo razo√°vel;
     - Analisar separadamente esse caso (ex.: "dia da falha").

O importante √©:

> N√£o tratar outliers automaticamente.  
> Primeiro, tente entender **por que** aquele valor est√° ali.
